```python
pairs = input.split(',')
segments = []

for p in pairs:
    x,y = p.split('-')
    segments.append((int(x), int(y)))
```

# PART 1

```python
total = 0
for start,end in segments:
    current = start
    while current <= end:
        str_num = str(current)
        len_num = len(str_num)
        if len_num % 2 == 0:
            first_half_start = current // 10**(len_num // 2)

            if 10**len_num > end:
                first_half_end = end // 10**(len_num // 2)
            else:
                first_half_end = 10**(len_num // 2) - 1
                                
            for to_double in range(first_half_start, first_half_end + 1):
                sim = to_double * 10**(len_num // 2) + to_double

                if start <= sim <= end:
                    total += sim

        current = 10**(len_num)

print(f'Solution: {total}')
```

# PART 2

```python
def get_divisors(n):
    divisors = []
    for i in range(1, n + 1):
        if n % i == 0:
            divisors.append(i)
    return divisors
```

```python
def check_valid(id:int):
    id_len = len(str(id))
    s_id = str(id)
    potential_lens = get_divisors(id_len)[:-1]

    for d in potential_lens:
        index = 0
        valid = True
        while index <= id_len-2*d and valid:
            valid = s_id[index:index+d] == s_id[index+d:index+2*d]
            index+=d
        if valid and index > 0:
            return id
    
    return 0
```

```python
counter = 0
for seg in segments:
    for potential_id in range((seg[0]), (seg[1]+1)):
        counter +=check_valid(potential_id)

print(f'Solution: {counter}')
```
