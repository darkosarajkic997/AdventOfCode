# INPUT

```python
ranges = []
ingredients = []
ranges_end = False
with open('input.txt') as f:
    for line in f:
        if not ranges_end:
            if line == '\n':
                ranges_end = True
            else:
                start, end = line.strip().split('-')
                ranges.append([int(start),int(end)])
        else:
            ingredients.append(int(line.strip()))
```

# PART 1

```python
counter = 0
for ing in ingredients:
    for r in ranges:
        if ing >= r[0] and ing<=r[1]:
            counter+=1
            break

print(f'Solution: {counter}')
```

# PART 2

```python
sorted_ranges = []

for start,end in ranges:
    added = False
    index = 0
    while not added and index < len(sorted_ranges):
        if start > sorted_ranges[index][0]:
            index+=1
        else:
            sorted_ranges.insert(index,[start,end])
            added = True
    if not added:
        sorted_ranges.append([start, end])
```

```python
current_end = sorted_ranges[0][1]
current_start = sorted_ranges[0][0]

total = current_end - current_start + 1

for start,end in sorted_ranges[1:]:
    if start <= current_end and end > current_end:
        total+=end - current_end
        current_end = end
    if start > current_end:
        current_end = end
        total+=end-start+1

print(f'Solution: {total}')
```
