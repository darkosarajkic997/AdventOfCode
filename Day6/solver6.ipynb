# PART 1

```python
total  = 0

for index in range(0,len(inputs[0])):
    if inputs[4][index] == '*':
        total+=int(inputs[0][index]) * int(inputs[1][index]) * int(inputs[2][index]) * int(inputs[3][index])
    else:
        total+=int(inputs[0][index]) + int(inputs[1][index]) + int(inputs[2][index]) + int(inputs[3][index])

print(f'Solution: {total}')
```

# PART 2

```python
inputs = []
with open('input.txt') as f:
    for line in f:
        inputs.append(line.strip('\n')+' ')
```

```python
def convert_ch_to_int(numbers):
    number = 0
    for num in numbers:
        if num != ' ':
            number = number*10 + int(num)
    
    return number
```

```python
#aligne input

max_len = max(len(inputs[0]),len(inputs[1]),len(inputs[2]),len(inputs[3]),len(inputs[4]))
max_len

index = 0

while index < len(inputs):
    diff = max_len - len(inputs[index])
    inputs[index]+=' '*diff
    index+=1
```

```python
total = 0
index = 0
current_op = '*'
perform_op = False
num_buffer = []

while index < max_len:
    num1 = inputs[0][index]
    num2 = inputs[1][index]
    num3 = inputs[2][index]
    num4 = inputs[3][index]
    op = inputs[4][index]

    if op != ' ':
        current_op = op

    if num1 == num2 == num3 == num4 == op == ' ':
        tmp = 1  if current_op == '*' else 0
        for num in num_buffer:
            tmp = tmp * num if current_op == '*' else tmp + num

        total+=tmp
        num_buffer = []
    else:
        num_buffer.append(convert_ch_to_int([num1, num2, num3, num4]))

    index+=1

print(f'Solution: {total}')

```
