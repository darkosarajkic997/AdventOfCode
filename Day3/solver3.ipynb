# INPUT

```python
with open('input.txt','r') as f:
    numbers = f.readlines()
```

# PART 1

```python
total = 0
for jolt in numbers:
    jolt = jolt.strip()
    len_j = len(jolt)
    max = 0
    max_next = 0
    for index in range(0,len_j-1):
        current = int(jolt[index])
        current_next =  int(jolt[index + 1])
        if current > max:
            max = current
            max_next = current_next
        elif current_next > max_next:
            max_next = current_next

    total += 10*max + max_next


print(f'Solution: {total}')
```

# PART 2

```python
total = 0

for jolt in numbers:
    seq_len = 12
    jolt = jolt.strip()
    len_j = len(jolt)
    finished = False
    start_search_index= 0
    end_search_index = len_j - seq_len + 1
    final_number = 0
    while not finished:
        final_number*=10
        local_max = 0
        local_max_id = 0
        for index in range(start_search_index,end_search_index):
            loc = int(jolt[index])
            if  loc > local_max:
                local_max = loc
                local_max_id = index
            if local_max == 9:
                break    
        final_number += local_max
        start_search_index = local_max_id + 1
        seq_len-=1
        end_search_index = len_j - seq_len + 1

        if seq_len==0:
            finished = True
            total+=final_number

print(f'Solution: {total}')
```
