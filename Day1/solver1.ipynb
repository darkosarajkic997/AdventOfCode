```python
with open('input.txt', 'r') as file:
    content = file.readlines()
```

```python
# PART 1

counter = 0 
start = 50
for entry in content:
    sgn = 1 if entry[0] == 'R' else -1
    value = int(entry.strip()[1:])

    start = (start + sgn * value) % 100

    if start == 0:
        counter+=1


counter
```
